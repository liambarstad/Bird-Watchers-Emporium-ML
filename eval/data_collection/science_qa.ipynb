```python
import pandas as pd
from datetime import datetime
from tqdm import tqdm


def classify_question(question: str, model_name: str):

	instructions = '''
You are a classifier, with the task of determining whether or not a particular question pertains to a bird or the discipline of bird watching.
Respond ONLY with the word "YES" if it does pertain to a bird or bird watching, or "NO" if it does not.'''
	
	curr_try = 1
	while curr_try <= 3:
		is_bird_resp = ollama.chat(
			model=model_name,
			messages=[{
				'role': 'system', 
				"content": instructions
				},{ 
				'role': 'user',
				'content': question
			}],	
		)

		is_bird	= is_bird_resp['message']['content'].split('\n\n')[-1]

		if is_bird != 'YES' and is_bird != 'NO':
			print(f'{datetime.now()}:: Try: {curr_try}, Found non-classifed output: **{is_bird.replace('\n', ' ')}** For Question: **{question}**')
		elif is_bird == 'YES':
			return 1
		elif is_bird == 'NO':
			return 0
		curr_try += 1
		
	raise ValueError(f'model {model_name} Not Classified')
    
    
class_df_rows = []

for ind, row in tqdm(val_df.iterrows()):

	is_bird_qwen = classify_question(row.question, 'qwen3:14b')
	is_bird_gpt = classify_question(row.question, 'gpt-oss:20b')

	class_df_rows.append({
		'is_bird_qwen': is_bird_qwen,
		'is_bird_gpt': is_bird_gpt,
		**row
	})
	
	print(f'{datetime.now()}:: Classified Question: **{row.question}** GPT - {is_bird_gpt}, QWEN - {is_bird_qwen}')


class_df = pd.DataFrame(class_df_rows)
```
