```python
from langchain_core.messages import (
    SystemMessage,
    AIMessage
)

questions = [
	#'What are the closest blue birds near me?',
	'What do I need to go to see an Ostrich?',
	#'What are some easy birds to start off with in my area?',
	'What are some easy birds to start off with?',
	'How many red birds are there in Australia?',
	'What equipment will I need to start bird watching?'
]

initial_messages = [
	SystemMessage(content=(
		'You are a friendly agent for the Bird Watcher\'s Emporium. '
		'Your goal is to provide information to the user about birds they might want to watch, '
		'get them excited about the hobby of bird watching, '
		'and sell them goods and services that can assist them in their bird watching endeavors. '
	)), 
 	AIMessage(content=(
		'Welcome to Bird Watchers\' Emporium! '
		'I can help you with anything and everything you\'ll need to start finding and identifying cool birds in your area or abroad! '
		'Need help identifying what kind of birds you want to see? I can accept descriptions, locations, or other specifications so we can narrow down your choices. '
		'Once we\'ve decided what birds you want to add to your field list, I can help you plan your trip and find any necessary equipment you\'ll need to begin your bird watching adventure! '
	))
]
```

```python
from langchain_ollama import ChatOllama

llm = ChatOllama(
	model='gpt-oss:20b',
	temperature=0.2,
	base_url='http://127.0.0.1:11434'
)

for q in questions:
	print('------------------------------------------')
	print(f'Q: {q}')
	resp = llm.invoke([*initial_messages, q])
	print('-----')
	print(f'A: {resp.content}')
```
